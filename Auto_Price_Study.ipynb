6) Использовать библиотеку statsmodels для построения линейной регрессии с одним предиктором – horsepower.

Рассчитать процент изменчивости (R^2) для полученной модели и округлить до целого числа.

```python
import pandas as pd
import statsmodels.formula.api as smf 

df = pd.read_csv('cars.csv')

df['company'] = df.CarName.str.split(' ').apply(lambda x: x[0])
df = df.drop(['car_ID', 'CarName'], axis=1)

df['company'] = df.company.str.lower()
df['company'] = df.company.replace({'maxda': 'mazda', 'porcshce': 'porsche', 'toyouta': 'toyota', 'vokswagen': 'volkswagen', 'vw': 'volkswagen'})

df = df[['price', 'company', 'fueltype', 'aspiration','carbody', 'drivewheel', 'wheelbase', 'carlength','carwidth', 'curbweight', 'enginetype', 'cylindernumber', 'enginesize', 'boreratio','horsepower']]

result = smf.ols('price ~ horsepower', data=df).fit()

print(result.summary())
```

Модель с одним предиктором horsepower объясняет примерно 65% изменчивости в цене автомобилей. Это говорит о том, что мощность двигателя значительно влияет на стоимость автомобиля.

7) Использовать библиотеку statsmodels для построения линейной регрессии с одним предиктором – horsepower.

Построение модели со всеми предикторами:

Преобразовать категориальные переменные в фиктивные переменные (dummy variables).Использовать все доступные предикторы для построения модели.

Построение модели со всеми предикторами, кроме марок машин:

Исключить предикторы, относящиеся к маркам машин, и построить модель.

Оценка показателей всех трех моделей:

Вывести сводку результатов для каждой модели, включая R-squared, коэффициенты и значимость предикторов.

```python
import pandas as pd
import statsmodels.api as sm

df = pd.read_csv('cars.csv')

df['company'] = df.CarName.str.split(' ').apply(lambda x: x[0])
df = df.drop(['car_ID', 'CarName'], axis=1)

df['company'] = df.company.str.lower()
df['company'] = df.company.replace({'maxda': 'mazda', 'porcshce': 'porsche', 'toyouta': 'toyota', 'vokswagen': 'volkswagen', 'vw': 'volkswagen'})
df = df[['price', 'company', 'fueltype', 'aspiration','carbody', 'drivewheel', 'wheelbase', 'carlength','carwidth', 'curbweight', 'enginetype', 'cylindernumber', 'enginesize', 'boreratio','horsepower']]
df_dummy = pd.get_dummies(data=df[['company', 'fueltype', 'aspiration','carbody', 'drivewheel', 'wheelbase', 'carlength','carwidth', 'curbweight', 'enginetype', 'cylindernumber', 'enginesize', 'boreratio','horsepower', 'price']], drop_first = True)


X = df_dummy.drop(columns=['price'])
Y = df_dummy['price']  
X = sm.add_constant(X)  
model = sm.OLS(Y, X)  
results = model.fit()  
print(results.summary())

X = df_dummy.drop(columns=['price', 'company_audi', 'company_bmw', 'company_buick', 'company_chevrolet', 'company_dodge', 'company_honda', 'company_isuzu', 'company_jaguar', 'company_mazda', 'company_mercury', 'company_mitsubishi', 'company_nissan', 'company_peugeot', 'company_plymouth', 'company_porsche', 'company_renault', 'company_saab', 'company_subaru', 'company_toyota', 'company_volkswagen', 'company_volvo'])
Y = df_dummy['price']  
X = sm.add_constant(X)  
model = sm.OLS(Y, X)  
results = model.fit()  
print(results.summary())

```

Модель с одним предиктором (horsepower):

R^2=0.653: Мощность двигателя (horsepower) объясняет примерно 65% изменчивости цены автомобилей.

Модель со всеми предикторами:

R^2=0.959: Модель, включающая все предикторы, в том числе и марки машин, хорошо объясняет более 95% изменчивости цены.

Модель со всеми предикторами, кроме марок машин:

R^2=0.914: Этот вариант модели, исключая марки машин, также обладает высоким R^2, равным примерно 91%.


Таким образом, можно сделать вывод, что технические характеристики автомобилей, особенно мощность двигателя, существенно влияют на цену.
